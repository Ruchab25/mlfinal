```python
#Measure of Variability
print(temp['Citations per Document'].var())
print(temp['Citations per Document'].std())

# mean (average) absolute deviation
print(temp['Citations per Document'].mad())

#range
range1 = max(temp['Citations per Document'])-min(temp['Citations per Document'])
print(range1)

#IQR
Q1 = np.percentile(temp['Citations per Document'], 25)
Q3 = np.percentile(temp['Citations per Document'], 75)
print(Q1, Q3)
IQR = Q3 - Q1
print(IQR)
```

```python
Q1 = np.percentile(temp['Citable documents'], 25.) # 25th percentile of the data of the given feature
Q3 = np.percentile(temp['Citable documents'], 75.) # 75th percentile of the data of the given feature
IQR = Q3-Q1 #Interquartile Range
ll = Q1 - (1.5*IQR)
ul = Q3 + (1.5*IQR)
upper_outliers = temp[temp['Citable documents'] > ul].index.tolist()
lower_outliers = temp[temp['Citable documents'] < ll].index.tolist()
bad_indices = list(set(upper_outliers + lower_outliers))
drop = True
if drop:
    temp.drop(bad_indices, inplace = True, errors = 'ignore')
    print('Outliers from {} feature removed'.format(temp['Citable documents']))
```
